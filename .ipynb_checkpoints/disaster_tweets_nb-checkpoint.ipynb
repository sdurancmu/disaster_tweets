```python
import pandas as pd
```

```python
train_data = pd.read_csv('train.csv')
train_data.columns
```

```python
for i in range(50,60):
    print(train_data.text.iloc[i])
    print(train_data.target.iloc[i])
```

```python
def clean_text(text):
    text = ' '.join(['~URL~' if 'http://' in i or 'https://' in i else i for i in text.split(' ') ])
    text = ' '.join(['~@user~' if len(i)>0 and i[0]=='@' else i for i in text.split(' ') ])
    return text

clean_text = [clean_text(i) for i in train_data.text]
```
